## Numpy

- [a] -> 1-D array

- [ [a], [b] ] -> 2-D array

- [ [ [a], [b], [c] ], [ [d], [e], [f] ] ] -> 3-D array

```python
my_matrix = [[[1,2,3], [4,5,6], [7,8,9]], [[1,2,7], [2,3,5], [3,4,6]]]
my_matrix
```

```python
np.array(my_matrix)
```

###

- arange function provides array with values from starting index to (ending index - 1) 

        ex: np.arange(1,10) -> Gives elements from 1 to 9

- It may contain a 3rd arguement with step size

```python
np.arange(0, 10)
```

```python
np.arange(1, 11, 2)
```

```python
np.arange(1, 2, 0.1)
```

###

- zeros function will give an array of zeros; zeros will be floating 

        ex: np.zeros(10) -> Gives array of 10 floating zeros

- can also give 2-D or 3-D arrays

- np.zeros((no. of arrays, rows, columns))

```python
np.zeros(10)
```

```python
np.zeros((3, 3))
```

```python
np.zeros((5,5,5))
```

```python
np.zeros((3,2,4))
```

###

- np.ones is same as np.zeros

```python
np.ones((4,3,5))
```

###

- np.linspace(x,y,z) finds z elements that are evenly spaced between x and y

        ex: np.linspace(0,10,3) -> 3 evenly spaced elements b/w 0 and 10

```python
np.linspace(0,10,3)
```

```python
np.linspace(0,10,50)
```

###

- eye gives an identity matrix

        ex: np.eye(4) -> Gives a 4x4 identity matrix

```python
np.eye(4)
```

###

- Random numbers are also called pseudo random numbers

- rand gives any random numbers b/w 0 and 1

- It can give 1-D, 2-D.... array

        ex: np.random.rand(2) -> 2 random numbers b/w 0 and 1

```python
np.random.rand(2)
```

```python
np.random.rand(5,5)
```

###

- randn -> gives you random numbers centered around zero with higher chances of being close to zero; numbers b/w -1 and 1 are given with common mean and mod

- randn generates random numbers each time with different seed; seed functiion provides a set of numbers for generating

```python
np.random.seed(100)
```

```python
np.random.randn(2)
```

```python
random_array = np.random.randn(3, 4)
print(random_array)
```

###

- np.random.randint gives some random numbers b/w given numbers

        ex: np.random.randint(1,100,10) -> 10 random integers b/w 1 and 100

```python
np.random.randint(1,100,10)
```

```python
mynum = np.random.randint(1,100,25)
```

```python
mynum
```

###

- reshape returns the array with same data but after reshapping it

        ex: mynum = np.random.randInt((1,100,25))
            mynum.reshape(5,5) -> Gives a 5x5 2-D array of numbers in mynum 

```python
mynum.reshape(5,5)
```

```python
mynum2 = np.random.randint(1,100,16)
```

```python
mynum2.reshape(2,2,4)
```
