## Map

## Filter

## Reduce

- These three are high level functions

- Map works by applying the specified operation on every iterable

- Filter function checks if the specified function can be applied on the iterables or not

- Syntax: 

            name(function, iterable)

```python
b = [1,2,3,4,5,6,7,8,9,0,np.nan,np.nan]
```

```python
list(map(math.sqrt,b))                
```

```python
mapped = list(map(math.isnan,b))
```

```python
mapped
```

```python
sum(mapped)
```

```python
sum(list(map(math.isnan,b)))              # returns the sum of all false values in b; false values are nan
```

```python
list(filter(math.isnan,b))               # returns all null values from a list
```

```python
if len(list(filter(math.isnan,b)))>0:
    print("Missing values")
```

```python
if len(list(filter(not(math.isnan),b)))>0:             # Gives error because not is an operator but not a function here
    print("Missing values")
```

Fix for above error

```python
if len(list(filter(lambda x: not(math.isnan(x)),b))):                  # Gets fixed because lambda function makes the next lines as an anonymous function themselves
    print("Fixed")
```

- Reduce function reduces the iterables to a single value by applying a single operation

- The data type of output will be same as that of inputs

```python
from functools import reduce
```

```python
reduce(lambda x,y:(x+y),b)
```

```python
arr = [1,2,3,4,5,6,7,8,9]  

def ReturnSum(m,n):                    # This function will work in place of lambda function
    return m+n

# As lambda ->  lambda m,n:m+n 
```

```python
reduce(ReturnSum,arr)
```
