- Big data vs High Dimensional Data

- High Dimensional Data as well as Big Data are structured data i.e. they are labelled

- Example of big data:
        
        - Bank details

- If number of features (Columns) in data becomes high (>100) but rows are less, it becomes high dimensional data

- If number of features as well as rows are high (Millions), it becomes big data

- Big data is taken from a number of sources

- DataFrame is basically a collection of arrays i.e. Each row or column is seen as a 1-D Array

- This Dataframe contains multiple 1-D arrays forming 2-D arrays which are further called series

- In case of list, ndim gives output one because it already has one row and column

```python
e = 1
f = [1]
g = [[1]]
h = [[[1]]]

i = np.array(e)                # Gives 0-D array; No rows and columns are present
j = np.array(f)                # Gives 1-D array
k = np.array(g)                # Gives 2-D array
l = np.array(h)                # Gives 3-D array
```

```python
print(i.ndim)
print(j.ndim)
print(k.ndim)
print(l.ndim)
```

### Slicing

```python
nparray[1:3, 1:3]               # (Row index:Column index), (element/row index:ending index for next row) 
```

```python
slice1 = nparray[1:3, 1:3]            
```

```python
slice1[1,1]                     # Accessing elements of slice1
```

```python
slice1[1,1] = 90                # Changing elements of slice1
```
