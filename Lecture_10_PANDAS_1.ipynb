###

### Data:

        1. Qualitative  -  Non Numerical
        2. Quantitative  -  Numerical

###

### Qualitative:
    
            1. Nominal
            2. Ordinal



1. Nominal: 

            - Named i.e. can not be put in an order / Can't be ordered
            - ex: Gender, FuelType, Name, Address

2. Ordinal:

            - In a natural order i.e. can not be rearranged
            - Already present in an order
            - ex: Grade, Rank, Days

###

### Quantitative:

            1. Discrete
            2. Continuous

1. Discrete:

            - Any column that contains numbers in the form of intergers
            - Values displayed / visualised using bar graphs
            - Ex: 23, 45, 67, 34, 35 etc...

2. Continuous:

            - Any column that contains numbers in the form of real values
            - Values displayed / visualised using histogram
            - Ex: 23.45, 67.78 etc...
