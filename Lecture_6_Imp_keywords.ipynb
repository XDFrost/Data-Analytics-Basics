## Important Keywords

- sep : sep/separator arguement takes the arguemnt that separates 2 values

- names : names arguement gives the names to the columns

- index : index arguement provides names to the rows

- iloc : iloc provides the data for the mentioned "row index"; accepts numeric values

- loc : loc provides the data for the index "name"; accepts string
