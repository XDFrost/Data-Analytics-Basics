## Conditional Operators

- And operator -> &

- Or operator -> |

- Not operator -> ~

- Will give bool values

```python
a = [1,2,3,4,5]
a = np.array(a)
```

```python
a
```

```python
a>2
```

###

- AND (&)

```python
(a>2) & (a<5)
```

```python
a[(a>2) & (a<5)]
```

###

- OR (|)

```python
a[(a>2) | (a<5)]
```

###

- NOT (~)

```python
print((a>2))
print(~ (a>2))
```

## Arithmetic Operators

    -> +
    -> -
    -> *

    -> /

```python
arr = np.arange(0,10)
```

```python
arr + arr
```

```python
arr * arr
```

```python
arr - arr
```

```python
arr**3
```

```python
arr/arr                # Will give warning when division by 0, but won't give error; Replaces by nan
```

```python
1/arr                  # Gives warning but not an error; gives inf when we get 1/0
```
